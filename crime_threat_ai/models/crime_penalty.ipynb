```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/processed_crime_data.csv")
print(df.shape)
```

```python
severity_map = {
    'MURDER': 5,
    'RAPE': 5,
    'ROBBERY': 4,
    'SHOOTING': 5,
    'STABBING': 4,
    'CARNAPPING': 3,
    'MOTORNAPPING': 3,
    'PHYSICAL INJURIES': 3,
    'MAULING': 3,
    'THEFT': 1,
    'ACTS OF LASCIVIOUSNESS': 4
}

df['crime_severity'] = df['offense'].str.strip().str.upper().map(severity_map)
```

```python
def create_crime_penalty(df):
    crime_component = 0.4 * df['crime_severity']
    casualty_component = 0.3 * (df['victimKilled_binary'] + df['victimInjured_encoded'])

    df['crime_penalty'] = (
        crime_component + 
        casualty_component
    ).clip(0,1)

    return df
    
```

```python
create_crime_penalty(df)
```
