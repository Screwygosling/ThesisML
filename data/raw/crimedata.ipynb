```python
import pandas as pd
```

```python
df = pd.read_csv("crime_data.csv")
print(df.head())
```

```python
df = df.dropna(axis=1, how="all")

print(df.columns)
```

```python
df = df[[
    "dateCommitted", #
    "timeCommitted", #
    "barangay", #
    "municipal", #
    "lat", #
    "lng", #
    "offense", #
    "offenseType", #
    "incidentType", #
    "typeofPlace", #
    "victimCount", #
    "suspectCount", #
    "victimKilled", #
    "victimInjured", #
]]

print(df.head())
```

```python
df = df.copy()
df["dateCommitted"] = pd.to_datetime(df["dateCommitted"])
```

this converts date committed into individual months, time, and day of the week


```python
df["month"] = df["dateCommitted"].dt.month
df["day_of_week"] = df["dateCommitted"].dt.dayofweek
df["hour"] = df["dateCommitted"].dt.hour
```

```python
df = pd.get_dummies(df, columns=[
    "offenseType",
    "incidentType",
    "typeofPlace"
])
```

```python
crime_counts = df.groupby("barangay").size()

df["areaCrimeCount"] = df["barangay"].map(crime_counts)
```

```python
df["barangay_encoded"] = df["barangay"].astype("category").cat.codes
df["municipal_encoded"] = df["municipal"].astype("category").cat.codes
```

```python
df["lat_encoded"] = df["lat"].astype("float64")
df["lng_encoded"] = df["lng"].astype("float64")
```

```python
df["victimCount_encoded"] = pd.to_numeric(df["victimCount"],errors="coerce").fillna(0)
df["suspectCount_encoded"] = pd.to_numeric(df["suspectCount"],errors="coerce").fillna(0)
df["victimKilled"] = pd.to_numeric(df["victimKilled"], errors='coerce').fillna(0)
df["victimKilled_binary"] = (df["victimKilled"] > 0).astype(int)
df["victimInjured_encoded"] = pd.to_numeric(df["victimInjured"], errors="coerce").fillna(0)
```

```python
df.to_csv("processed_crime_data.csv", index=False)
print("Dataframe saved to: processed_crime_data.csv")
```
